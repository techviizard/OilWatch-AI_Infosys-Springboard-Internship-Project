```python
#Verifying images and masks count
import os

def verify_dataset(images_dir, masks_dir):
    images = sorted([f for f in os.listdir(images_dir) if f.endswith('.png')])
    masks = sorted([f for f in os.listdir(masks_dir) if f.endswith('.png')])

    print(f"Images count: {len(images)}")
    print(f"Masks count: {len(masks)}")

    image_ids = set([img.split('.')[0] for img in images])
    mask_ids = set([msk.split('.')[0] for msk in masks])

    missing_masks = image_ids - mask_ids
    missing_images = mask_ids - image_ids

    if missing_masks:
        print("❌ Images without masks:", missing_masks)
    if missing_images:
        print("❌ Masks without images:", missing_images)

    if not missing_masks and not missing_images:
        print("✅ Dataset verified: all images have matching masks")

# Paths
verify_dataset(
    "oil_spill_dataset/train/images",
    "oil_spill_dataset/train/masks"
)

verify_dataset(
    "oil_spill_dataset/val/images",
    "oil_spill_dataset/val/masks"
)
```

```python
#image mask pair for training
import os
import random
import shutil

train_img_dir = "oil_spill_dataset/train/images"
train_mask_dir = "oil_spill_dataset/train/masks"

test_img_dir = "oil_spill_dataset/test/images"
test_mask_dir = "oil_spill_dataset/test/masks"

os.makedirs(test_img_dir, exist_ok=True)
os.makedirs(test_mask_dir, exist_ok=True)

images = [f for f in os.listdir(train_img_dir) if f.endswith('.png')]
random.shuffle(images)

test_size = int(0.15 * len(images))
test_images = images[:test_size]

for img in test_images:
    img_path = os.path.join(train_img_dir, img)
    mask_path = os.path.join(train_mask_dir, img)

    shutil.move(img_path, os.path.join(test_img_dir, img))
    shutil.move(mask_path, os.path.join(test_mask_dir, img))

print(f"✅ Moved {test_size} image-mask pairs to test set")
```

```python
#verifying training|validation|testing images with their respective masks

print("\nTRAIN SET")
verify_dataset(
    "oil_spill_dataset/train/images",
    "oil_spill_dataset/train/masks"
)

print("\nVALIDATION SET")
verify_dataset(
    "oil_spill_dataset/val/images",
    "oil_spill_dataset/val/masks"
)

print("\nTEST SET")
verify_dataset(
    "oil_spill_dataset/test/images",
    "oil_spill_dataset/test/masks"
)
```
